```python
import pandas as pd
import matplotlib.pyplot as plt
from venn import venn
```

```python
s1_links = pd.read_excel("MS_Annika_results/S1_deiso_top2_fp_1%_Crosslinks_nodecoy.xlsx")
s2_links = pd.read_excel("MS_Annika_results/S2_deiso_top2_fp_1%_Crosslinks_nodecoy.xlsx")
s3_links = pd.read_excel("MS_Annika_results/S3_deiso_top2_fp_1%_Crosslinks_nodecoy.xlsx")
```

```python
def barplot(labels, counts, colors, title):
    fig, ax = plt.subplots()
    ax.set_xlabel("Replicate")
    ax.set_ylabel("Crosslinks at 1% FDR")
    ax.set_title(title)
    bars = ax.bar(labels, counts, color = colors)
    ax.bar_label(bars)
    return plt
```

```python
barplot(["S1", "S2", "S3"], [s1_links.shape[0], s2_links.shape[0], s3_links.shape[0]], ["black", "orange", "grey"], "full_fasta");
```

```python
def get_crosslinks(df):
    xl = set()
    for i, row in df.iterrows():
        current_xl = "--".join(sorted([str(row["Sequence A"]).strip() + "_" + str(row["Position A"]).strip(), str(row["Sequence B"]).strip() + "_" + str(row["Position B"]).strip()]))
        xl.add(current_xl)
    return xl
```

```python
s1 = get_crosslinks(s1_links)
s2 = get_crosslinks(s2_links)
s3 = get_crosslinks(s3_links)
print(f"s1: {len(s1)}, s2: {len(s2)}, s3: {len(s3)}")
```

```python
venn(s1, s2, s3, labels = ["S1", "S2", "S3"], colors = ["black", "orange", "grey"], title = "full_fasta");
```

```python
f"Total crosslinks at 1% FDR across S1, S2 and S3: {len(s1.union(s2, s3))}"
```
